# Change of Variables

Let $X$ be a continuous random variable with PDF $f_X$, and let $Y = g(X)$, where $g$ is a differentiable and strictly monotonic function.

When, $g(x)$ is strictly increasing, the CDF is given by,
$$
\begin{align*}
F_Y(y) &= P(Y \le y) \\
&= P(g(X) \le y) \\
&= P(X \le g^{-1}(y)) \\
&= F_X(g^{-1}(y)) \\
&= F_X(x) \tag{1}
\end{align*}
$$

Differentiating $(1)$ w.r.t. $x$, we get,
$$
f_Y(y) = f_X(x) \frac{dx}{dy} \tag{2}
$$

When, $g(x)$ is strictly decreasing, the CDF is given by,
$$
\begin{align*}
F_Y(y) &= P(Y \le y) \\
&= P(g(X) \le y) \\
&= P(X \ge g^{-1}(y)) \\
&= 1 - P(X \le g^{-1}(y)) \\
&= 1 - F_X(g^{-1}(y)) \\
&= 1 - F_X(x) \tag{3}
\end{align*}
$$

Differentiating $(3)$ w.r.t. $x$, we get,
$$
f_Y(y) = -f_X(x) \frac{dx}{dy} \tag{4}
$$

The strictly increasing and decreasing case can be combined together to form,
$$
f_Y(y) = f_X(x) \left\lvert\frac{dx}{dy}\right\rvert \tag{5}
$$

> **Theorem (Change of Variables in One Dimension)**
> 
> Let $X$ be a continuous random variable with PDF $f_X$, and let $Y = g(X)$, where $g$ is a differentiable and strictly monotonic function. Then,
> $$f_Y(y) = f_X(x) \left\lvert\frac{dx}{dy}\right\rvert$$

**On strict monotonicity.** What happens if $g$ is not a strictly monotonic function? Then, the function won't be invertible, i.e. different values of $x$ will get mapped to the same $y$. *Why is this a problem?*

## Example: $g(x) = e^{-x}$

Let $X \sim \mathcal{N}(0, 1)$, and $y = e^{-x}$.Then,
$$
\begin{align*}
f_Y(y) &= f_X(x) \left\lvert \frac{dx}{dy} \right\rvert \\
&= \varphi(x) \left\lvert \frac{-\ln{y}}{dy} \right\rvert \\
&= \varphi(x) \left\lvert -\frac{1}{y} \right\rvert \\
&= \frac{1}{y} \varphi(\ln y)
\end{align*}
$$

```python
import seaborn as sns
import numpy as np
```

```python
def normal(x, mean = 0, std = 1):
    return np.exp(- 1 / 2 * ((x - mean) / std) ** 2) / (std * (2 * np.pi) ** (1 / 2))
```

```python
n_samples = 1000
x = np.linspace(-5, 5, n_samples)
dist_x = normal(x)
```

```python
dist_x.sum() * 10 / n_samples # checking if it adds up to 1... 
```

```python
y = np.exp(-x)
```

```python
dist_y = 1 / y * normal(np.log(y))
```

```python
sns.lineplot(x=np.linspace(-3, 3, n_samples), y=dist_x[:])
sns.lineplot(x=np.linspace(-3, 3, n_samples), y=dist_y[:])
```
