# Gamma

The Gamma function is defined as,
$$
\Gamma(a) = \int_{0}^\infty x^{a} e^{-x} \frac{dx}{x}
\tag{1}
$$
where, $a > 0$ is a real number.

If we divide both sides of $(1)$ by $\Gamma(a)$, then we obtain
$$
1 = \int_0^\infty \frac{1}{\Gamma(a)}x^a e^{-x} \frac{dx}{x}
\tag{2}
$$

The integrand in $(2)$ looks like a PDF, 
$$
f_X(x) = \frac{1}{\Gamma(a)}x^a e^{-x} \frac{1}{x} \tag{3}
$$
This is the PDF of $X \sim \text{Gamma}(a, 1)$.

If we perform a scale transformation on $X$ then we can obtain the PDF for generalized Gamma distribution. Let $Y \sim \text{Gamma}(a, \lambda)$ such that $Y = X / \lambda$. Then, the PDF of $Y$ can be obtained through change of variables as,
$$
\begin{align*}
f_Y(y) &= f_X(x) \left\lvert\frac{dx}{dy}\right\rvert \\
&= f_X(x) \lambda \\
&= \frac{1}{\Gamma(a)}x^a e^{-x} \frac{1}{x} \lambda \\
&= \frac{1}{\Gamma(a)}(\lambda y)^a e^{-\lambda y}\frac{1}{\lambda y} \lambda \\
&= \frac{1}{\Gamma(a)}(\lambda y)^a e^{-\lambda y}\frac{1}{y}
\tag{4}
\end{align*}
$$

```python
import math
import seaborn as sns
import numpy as np
```

```python
def gamma(x, a, l):
    assert (x >= 0).all(), "gamma distribution only exists for x > 0"
    return (l * x) ** (a - 1) * np.exp(- l * x) * l / math.gamma(a)
```

```python
x = np.linspace(0, 20, 100)
```

```python
l = 0.5
f_X = gamma(x, 3, 1)
```

```python
f_Y_1 = gamma(0.5 * x, 3, 1) * l  # to show that f_X(x) * lambda also gives the same thing
```

```python
f_Y_2 = gamma(x, 3, l)
```

```python
sns.lineplot(x=x, y=f_X)
sns.lineplot(x=x, y=f_Y_1)
sns.lineplot(x=x, y=f_Y_2, linestyle="--")
```
